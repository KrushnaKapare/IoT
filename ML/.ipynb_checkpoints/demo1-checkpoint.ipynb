```python
def function3():
    arr = np.array([12,20,30,40,50,60])
    print(arr)
    print(f"type = {type(arr)}")
    print(f"Dtype = {arr.dtype}")
    print(f"ItemSize = {arr.itemsize}")
    print(f"size = {arr.size}")
    print(f"NDIM = {arr.ndim}")
    print(f"NBYTES = {arr.nbytes}")    
```

```python
function3()
```

```python
def printProperties(arr):
    print(arr)
    print(f"type = {type(arr)}")
    print(f"Dtype = {arr.dtype}")
    print(f"Item size = {arr.itemsize}")
    print(f"size = {arr.size}")
    print(f"NDIM = {arr.ndim}")
    print(f"NBYTES = {arr.nbytes}")
n1 = np.array([2,5,8,9,3,4,1,7,6,0,11,22,33,44,55])
printProperties(n1)
```

```python
n3 = np.array([11,22,33,44],dtype = np.int32)
printProperties(n3)
```

```python
n4 = np.array([1.15,2.32,4.3])
printProperties(n4)
```

```python
n5 = np.array(["C","c++","java","python"])
printProperties(n5)
```

```python
country = np.array(["India","USA","Australia"])
printProperties(country)
```

```python
a = np.array(["krushna"])
printProperties(a)
```
